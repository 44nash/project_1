```python
import pandas as pd
import numpy as np
import os
from dotenv import load_dotenv
# will need to run: pip install quandl
import quandl
import plotly.express as px
```

```python
#import consumer price data from quandl API. Imports directly as a DataFrame
CPI_USA = quandl.get("RATEINF/CPI_USA")
```

```python
CPI_USA
```

```python
# Convert CPI into inflation rate. Calculates the change in CPI (compared to the same month in the prior year). For example: March 2021 Inflation Rate = (March 2021 CPU )
inflation_rate = (CPI_USA - CPI_USA.shift(12))/CPI_USA
CPI_USA.shift(12)
```

```python
inflation_rate = inflation_rate.dropna()
inflation_rate.columns = ['Inflation Rate']
```

```python
plot = px.line(inflation_rate)
plot
```
